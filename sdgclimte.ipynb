```python
print("="*5)
print("Hypotheses testing")
print("="*5)

from scipy.stats import ttest_ind

# Is elevation significantly different between spruce/fir (type 1) and lodgepole pine (type 2)?
spruce = df[df['Cover_Type'] == 1]['Elevation']
lodgepole = df[df['Cover_Type'] == 2]['Elevation']

ttest_ind(spruce, lodgepole)
```
